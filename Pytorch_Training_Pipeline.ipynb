## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size=0.2)
```

# Scaling 

All teh values of dataframe are on different scale, it is best to scale them so that all the values lies within a range

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

As the labels are Letters not numbers, and our neural network only understand numbers, so we have to convert the labels to some numbers. For that we'll use `Label Encoder`

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
```

## Numpy arrays to Pytorch tensors

All the the dataframes (X_train, X_test, y_train & y_test) are based on numoy arrays. But as we have build the model in pytorch, so we have to convert them into Pytorch **tensors**

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

## Defining The Model

For starting we'll be creating a very simple Neuron, which have only weights and biases.

As our training dataset have 30 input features, so neuron will get 30 inputs. For our usecase these 30 values will be random.

```python
# Take the one index value of X_train shape, this actually indicates the
# total columns in X_train 
torch.rand(X_train_tensor.shape[1],1)
```

```python
from torch import nn

class SimpleANN(nn.Module):

    def __init__(self, X):
        self.weights = torch.rand(X.shape[1],1, dtype=torch.float64, requires_grad=True)
        self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)
        

    def forward(self,X):
      z = torch.matmul(X, self.weights) + self.bias
      y_pred = torch.sigmoid(z)
      return y_pred
    
    def loss_function(self, y_pred, y):
       # Calulate predictions to aviod log(0)
       epsilon = 1e-7
       y_pred = torch.clamp(y_pred, epsilon, 1- epsilon)

       # Calculate Loss
       loss = -(y_train_tensor * torch.log(y_pred) + (1- y_train_tensor) * torch.log(1-y_pred)).mean()
       return loss
```

### Important Parameters

```python
learning_rate = 0.1
epochs = 25
```

## Training Pipeline

Training Pipeline Contains Following Steps
- Creating Model
- Forward Pass
- Calculate Loss
- Backpropagation
- Parameter Update

All these steps happen in loop and loop runs for defined number of time as we have defined in `epochs`.
We'll each of these steps in detail

```python
# create Model
model = SimpleANN(X_train_tensor)

# define loop
for epoch in range(epochs):

    # forward pass
    y_pred = model.forward(X_train_tensor)

    # loss calculate
    loss = model.loss_function(y_pred, y_test)
    
    # backward pass
    loss.backward()

    # parameters update
    with torch.no_grad():
        model.weights -=learning_rate * model.weights.grad
        model.bias -=learning_rate * model.bias.grad

    # Zero gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_() 

    # print loss iin each epoch
    print(f'Epoch: {epoch}, Loss: {loss.item()}')
```

## Model Evaluation

```python
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f'Accuracy : {accuracy.item()}')
```
